### 1.2. Notebooks - Conceptos básicos

Un notebook se organiza en celdas que siguen un orden secuencial. En Google Colab, hay 2 tipos principales de celdas: **texto** (llamado *markdown*) y **código**. Ambos tipos de celdas se pueden ejecutar, pero tienen diferentes funciones. Una celda de tipo *código* se emplea para ejecutar el bloque de código de Python que esté dentro de esa celda. Por otra lado, una celda *markdown* nos permite dar formato al texto (por ejemplo, se pueden poner elementos en **negrita** o *cursiva*) o añadir imágenes en el notebook.

Para poder añadir celdas, se puede clicar las opciones "+ Codigo" o "+ Texto" de la parte superior izquierda, o se puede ir por Insertar > Celda de código/texto.
