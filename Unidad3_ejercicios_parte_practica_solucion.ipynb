### Ejercicio 1

En este ejercicio vamos a poner en práctica la funcionalidad de `break` y `continue`.

Dada la lista:

```python
lista = [True, "Python", "Hello World", 56 , 17, False, "Hola Mundo"]
```

a) ¿Qué diferencia hay entre utilizar `break` o `continue` en los apartados a1) y a2)? Razona la respuesta.

```python
#a1)

for element in lista:
    if element == "Python":
        break
    print(element)
```

```python
#a2)

for element in lista:
    if element == "Python":
        continue
    print(element)
```

```python
# Respuesta

#a1)

# Recorremos los elementos de la lista con un for
for element in lista:
    # Cuando el elemento sea "Python", parará el for debido al break
    if element == "Python":
        break
    # Mostramos el elemento por pantalla
    print(element)
```

```python
# Respuesta

# Recorremos los elementos de la lista con un for
for element in lista:
    # Cuando el elemento sea "Python"
    # pasará al siguiente elemento debido al continue
    if element == "Python":
        continue
    # Mostramos el elemento por pantalla
    print(element)
```

**Respuesta**

La diferencia entre los dos apartados reside en la funcionalidad de break y continue. Cuando usamos break, hacemos que el loop pare cuando se encuentra el string "Python", por tanto solo mostramos el primer elemento. En cambio, cuando usamos el continue, nos saltamos el string "Python" pero seguimos mostrando el resto de elementos de la lista.

b) Escribe un código que, dada la lista definida anteriormente, 1) detecte los elementos **Boolean** de la lista, y 2) los muestre por pantalla.

```python
# Respuesta

#b)

# Recorremos los elementos de la lista con un for
for element in lista:
    # En el caso que el tipo de elemento sea bool
    if type(element) == bool:
        # Mostramos el elemento por pantalla
        print(element)
```

c) Escribe un código que, mediante `if`, muestre por pantalla los string que empiezan por "H".

Para solucionar este problema tenéis que usar 2 condiciones:

- 2) Seleccionar los strings que empiezan por "H"

### Ejercicio 2

a) Las *list comprehensions* son muy útiles para crear listas en una sola línea de código. Mucho del código que escribimos con bucles for se podría adaptar a este formato más breve.

A continuación, te proponemos que transformes los códigos siguientes a *list comprehensions*:

```python
#a1)

result = []

for i in range(7):
    if i%2 == 0:
        result.append(i**2)

print(result)
```

```python
#a2)

list_elem2 = [10, 45, 57, 4, 2, 13, 98, 7]
new_list = []

for x in list_elem2:

    if x >= 45:
        new_list.append(x+1)
    else:
        new_list.append(x+5)

print(new_list)
```

```python
# Respuesta

result = [i**2 for i in range(7) if i%2 == 0]

# Mostramos el resultado por pantalla
print(result)
```

```python
# Respuesta

#a2)

# Creamos la lista
list_elem2 = [10, 45, 57, 4, 2, 13, 98, 7]

new_list = [x+1 if x >= 45 else x+5 for x in list_elem2]

# Mostramos el resultado por pantalla
print(new_list)
```

### Ejercicio 3

Para practicar las funciones, vamos a crear 3 funciones distintas, con las siguientes características:

a) Crea una función que, dada una palabra (por ejemplo, arbol), te muestre por pantalla el mensaje:

`La palabra arbol empieza por la letra A`.

La función tiene que mostrar y devolver la 1a letra de la palabra en mayúsculas.

**Nota**: Podéis usar la función [capitalize](https://docs.python.org/3/library/stdtypes.html#str.capitalize).

b) Crea una función que, dada una lista con nombres, ordene la lista alfabéticamente y devuelva el primer elemento. También se tiene que mostrar el siguiente mensaje:

`El primer nombre de la lista ordenada alfabéticamente es Carlos`


**Nota**: Podéis revisar la [documentación oficial](https://docs.python.org/es/3/howto/sorting.html) de Python para encontrar la función para ordenar una lista.

c) Crea un diccionario de 6 elementos con series de TV y su puntuación (del 0 al 10). A continuación, crea una función que, dado este diccionario, devuelva una lista con solo el nombre de las series que superen una determinada puntuación (por ejemplo, 7 o más). Este valor se debe especificar como input de la función también.

```python
## a) Función letra_inicio

# Definimos la función con palabra como variable input
def letra_inicio(palabra):
  # Nos quedamos con la 1a letra en mayúsculas
  letra = palabra[0].capitalize()
  # Mostramos el mensaje por pantalla
  print(f"La palabra {palabra} empieza por la letra {letra}")
  # Devolvemos la 1a letra como output
  return letra
```

```python
# Comprobamos que la función es correcta
letra = letra_inicio("sushi")
print("Output: ", letra)
```

```python
## b) Función primer nombre

def primer_nombre(lista_nombres):
  # Ordenamos la lista alfabéticamente
  lista_ordenada = sorted(lista_nombres)
  # Nos quedamos con el 1r nombre
  nombre1 = lista_ordenada[0]
  # Mostramos el mensaje por pantalla
  print(f"El primer nombre en orden alfabético es {nombre1}")
  # Devolvemos el 1r nombre como output
  return nombre1
```

```python
# Comprobamos que la función es correcta
lista_nombres_CSM = ["Makima", "Pochita", "Denji", "Aki", "Power"]
nombre1 =  primer_nombre(lista_nombres_CSM)
print("Output: ", nombre1)
```

```python
## c) Función top series

def series_top(dict_series, valor):
  # Creamos una lista vacía
  top_list = []
  # Iteramos por todos los elementos del diccionario
  for serie in dict_series:
    # Si la puntuación es igual o mayor al valor
    if dict_series[serie] >=valor:
      # Se añade la serie a la lista
      top_list.append(serie)
  # Devolvemos la lista
  return top_list
```

```python
# Creamos un diccionario con las series y su puntuación
dict_series = {
    'Juego de tronos': 8,
    'The Witcher': 6,
    'El cuento de la criada': 9,
    'Élite': 4,
    'Lupin': 7,
    'El cuerpo en llamas': 7,
    'One Piece': 8,
    'La casa de papel': 7,
    'Paquita Salas': 10,
    'Arcane': 9
}

# Especificamos un valor
valor = 8
# Comprobamos que la función es correcta
top_list = series_top(dict_series, valor)
# Mostramos el resultado por pantalla
print(f"Las series con una puntuación igual o mayor a {valor} son:\n {top_list}")
```

En este ejercicio vamos a crear una versión simplificada de una calculadora, que llamaremos **funcion_calculadora**.

a) Para el correcto funcionamiento de la calculadora, los siguientes puntos se tienen que cumplir:

- La función tiene que tener 3 inputs: los elementos `a` y `b` (los números que vamos a usar), y la `operación` que queremos hacer con ellos. Por ejemplo, a y b pueden ser 2 y 3, y la operación, la suma. Para simplificar, vamos a permitir 4 operaciones: suma, resta, multiplicación y división. La variable operación va tener el valor suma por defecto (predefinido).

- Si se especifica una operación que no está entre las 4 establecidas, el programa tiene que lanzar una `Exception` y mostrar un mensaje de error.

- Como resultado, se tiene que guardar el resultado de la operación y también mostrar un mensaje por pantalla, por ejemplo:

`El resultado de la suma de 3 y 4 es 7!`


```python
# Respuesta

def funcion_calculadora(a, b, operacion = "suma"):
  # Definimos las operaciones posibles
  operaciones_cal = {"suma", "resta", "multiplicacion", "division"}

  # Si la opción no está predefinida en la lista, se lanza un mensaje de error
  if operacion not in operaciones_cal:
      raise Exception("[ERROR], la opción no es correcta! \n \
      Tienes que especificar una de las siguientes opciones: suma, resta, multiplicación o división")

  # Se realizan los cálculos según la operación elegida
  if operacion == "suma":
    c = a + b
  elif operacion == "resta":
    c = a - b
  elif operacion == "multiplicacion":
    c = a * b
  elif operacion == "division":
    c = a / b

  # Se muestra el resultado por pantalla
  print(f"El resultado de la {operacion} de {a} y {b} es {c}!")

  # Se devuelve el resultado de la operación
  return c
```

b) Ahora que tenemos una primera versión de funcion_calculadora, vamos a mejorarla.
- El usuario tiene que introducir los parámetros por pantalla, en vez de especificarlos en el código.

- Si el usuario introduce un elemento que no sea un número como variable, el programa tiene que avisar que es un valor erróneo y permitir al usuario que vuelva a introducir correctamente los parámetros. Esta opción tiene que permitirse solo una vez, si el usuario introduce 2 veces un valor incorrecto, el programa se cerrará.


```python
def funcion_calculadora(operacion="suma"):
  # Numero de intentos fallidos permitidos
  n = 2
  while n:
    try:
      print("Introduce el 1r número: ")
      a = int(input()) # Tenemos que convertirlo a int

      print("Introduce el 2º número: ")
      b = int(input()) # Tenemos que convertirlo a int

      print("Introduce la operación: ")
      operacion = input()

      # Si los valores están ok, se sale del while
      n = 0

      # Definimos las operaciones posibles
      operaciones_cal = {"suma", "resta", "multiplicacion", "division"}

      # Si la opción no está predefinida en la lista, se lanza un mensaje de error
      if operacion not in operaciones_cal:
          raise Exception("[ERROR], la opción no es correcta! \n \
          Tienes que especificar una de las siguientes opciones: suma, resta, multiplicación o división")

      # Se realizan los cálculos según la operación elegida
      if operacion == "suma":
        c = a + b
      elif operacion == "resta":
        c = a - b
      elif operacion == "multiplicacion":
        c = a * b
      elif operacion == "division":
        c = a / b

      # Se muestra el resultado por pantalla
      print(f"El resultado de la {operacion} de {a} y {b} es {c}!")

      # Se devuelve el resultado de la operación
      return c

    # Si los valores no son numéricos, se avisa al usuario
    except ValueError:
      # Mostramos el mensaje por pantalla
      n -=1
      print("[WARNING] Valor erróneo, sólo puede ser un número.")
```
