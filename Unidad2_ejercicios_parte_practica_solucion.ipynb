### Ejercicio 2

a) Crea una lista con los doce meses del año, de Enero a Diciembre.

b) Selecciona de la lista los meses impares y muéstralos por pantalla.

c) Muestra la lista de meses invertida (ordenada de Diciembre a Enero).

d) Crea las listas `primavera`,`verano`, `otoño`,` invierno`, que incluyan los meses correspondientes (e.g. primavera incluye los meses marzo, abril y mayo), a partir de los elementos de la lista de meses.

e) Crea un código que, dado un elemento `mes` (e.g. Enero) de la lista, muestre la frase siguiente:

"El mes del año que más me gusta es Enero".

f) Modifica el código del apartado e) para que el usuario introduzca el mes por pantalla.



````python
# Respuesta

# a) Creamos la lista de meses
meses = ["Enero","Febrero","Marzo","Abril","Mayo","Junio",
         "Julio","Agosto","Septiembre","Octubre","Noviembre","Diciembre"]

# b) Selecciona de la lista los meses impares y muéstralos por pantalla.
# Seleccionamos los meses impares
# Recuerda, python comienza con el índice 0!
print("Los meses impares son:", meses[0:11:2])
````

````python
# Respuesta

# c) Muestra la lista de meses invertida (ordenada de Diciembre a Enero).
# Mostramos la lista de meses al revés
print("La lista de meses invertida es la siguiente: ", meses[::-1])
````

````python
# Respuesta

# d) Crea las listas primavera, verano, otoño, invierno.
# Creamos las listas primavera, verano y otoño especificando los índices
primavera = meses[2:5]
verano = meses[5:8]
otoño = meses[8:11]

# En el caso del invierno, como los elementos no son consecutivos, primero creamos una lista
# que contiene el mes de Diciembre y, continuación, añadimos los meses de Enero y Febrero
invierno = [meses[11]]
invierno += meses[0:2]

# Mostramos los resultados por pantalla
print("Primavera: ", primavera)
print("Verano: ", verano)
print("Otoño: ", otoño)
print("Invierno: ", invierno)
````

````python
# Respuesta

# e)
# Seleccionamos el 6º mes, Junio
mes = meses[5]

# Utilizamos %s ya que queremos sustituir un string
print("El mes del año que más me gusta es %s" % (mes))
````

````python
# Respuesta

# f)
# El usuario introduce el mes por pantalla
print("Introduce tu mes favorito: ")
mes = input()

# Utilizamos %s ya que queremos sustituir un string
print("El mes del año que más me gusta es %s" % (mes.capitalize()))
````

### Ejercicio 5

A partir de la lista numeros = [1, 2, 3, 4, 5]:

a) Duplica el valor del último elemento de la lista y muestra la lista modificada por pantalla.

b) Crea una tupla a partir de la lista anterior.

c) Duplica el valor del primer elemento de la tupla y muestra la tupla modificada por pantalla.


¿Qué ocurre con la tupla? ¿Por qué? Razona la respuesta.

````python
# Creamos la lista que se especifica en el enunciado
numeros = [1, 2, 3, 4, 5]

# Duplicamos el último elemento de la lista
numeros[-1] = 2*numeros[-1]

# Mostramos la lista modificada por pantalla
print("Lista con elementos duplicados: ", numeros)

# Creamos la tupla con los elementos especificados
tupla_numeros = (1, 2, 3, 4, 5)

# Usamos try-except para capturar el error
# y que no se interrumpa la ejecución
try:
  # Duplicamos el primer elemento de la tupla
  tupla_numeros[0] = 2*tupla_numeros[0]
except:
  print("[ERROR] No se puede modificar la tupla")
````

En el caso de la tupla, nos encontramos un error debido a su característica de inmutabilidad, es decir, una vez declarada, una tupla no puede modificarse.
