### 1. Estructuras de control: if/while/for <a name="id1"></a>

#### 1.1. ¿Que son las estructuras de control?

En programación, un código nos indica una secuencia de instrucciones que el programa va a seguir. En general, estas instrucciones se van ejecutando siguiendo el orden especificado. Es decir, primero se ejecutan las instrucciones del principio, luego las siguientes, hasta llegar a la últimas. Cuando llegamos al final, el programa acaba.

Sin embargo, en algunos casos, vamos a querer que algunas instrucciones solo se ejecuten dada alguna condición, o vamos a querer que se repita alguna parte del código.

Para poder lograr esta funcionalidad, vamos a usar **las estructuras de control**. Estas estructuras se pueden separar en dos grandes grupos: los **condicionales** y los **bucles**.

Dentro de los condicionales, las estructuras básicas son el `if`, el `elif` y el `else`. Y en los bucles, el `for` y el `while`.

A continuación vamos a profundizar más en cada una de ellas.

**Estructuras condicionales: if-elif-else**

Las estructuras condicionales if-elif-else se usan para **tomar decisiones**, basándose en una o más condiciones.

Vamos a ver unos ejemplos:

```python
# Declaramos una variable a
a = 7

# Definimos una condición
if a >= 3:
  print("La variable a es igual o mayor que 3")
```

En este primer ejemplo, hemos puesto como condición que a sea mayor o igual que 3. Como a = 7, entonces la condición es cierta (se obtiene el boolean True), y se muestra el mensaje por pantalla.

¿ Que pasaría si la condición no se cumple?

```python
# ¿ Que pasaría si la condición no se cumple?
# Declaramos una variable a
a = 2

# Definimos una condición
if a >= 3:
  print("La variable a es igual o mayor que 3")
```

Como podemos ver, al no cumplirse la condición (ahora a es menor a 3), no entramos dentro del if, y, por lo tanto, no se muestra ningún mensaje.

Si queremos que haya algún mensaje por defecto si no se cumple la condición, podemos añadir el `else`.

```python
# ¿ Que pasaría si la condición no se cumple?
# Declaramos una variable a
a = 2

# Definimos una condición
if a >= 3:
  print("La variable a es igual o mayor que 3")
else:
  print("La variable a es menor que 3")
```

En este caso, si que se muestra un mensaje. Cuando no se cumple la condición del `if`, salta al `else`.

También hay la posibilidad que haya más de una condición. Para indicarlo, vamos a usar la estructura `elif`.

```python
# Declaramos dos variables, a y b
a = 2
b = 3

# Queremos que, dependiendo de la condición
# se muestre un mensaje diferente por pantalla

# Definimos la 1a condición con if
if a > b:
  print("La variable a es mayor a b")
# Definimos la 2a confición con elif
elif a < b:
  print("La variable a es menor a b")
# Si no se cumplen ninguna de las anteriores
else:
  print("La variable a es igual a b")
```

Como podemos ver en el código anterior, tenemos dos variables, a y b. En este caso, hemos escrito 3 mensajes diferentes, dependiendo de las condiciones.

Si inicializamos a con un valor de 2 y b con un valor de 3, entonces la condición que se cumplirá será la primera. Por este motivo, vemos que se muestra por pantalla el primer mensaje.

**Importante**: En python, para indicar que una sentencia está dentro de una estructura de control se usa el tabulador. Si una parte del código no está bien indentada, no va a funcionar de la manera esperada.

```python
# Declaramos una variable a
a = 5

# Definimos una condición
if a >= 3:
  print("La variable a es igual o mayor que 3")
else:
  print("Opción por defecto: ")
# Parte mal indentada, siempre se ejecuta
print("La variable a es menor que 3")
```

**Bucles for y while**

Los bucles `for` y `while` son **estructuras de control iterativas** que nos permiten repetir la ejecución de partes del código según unas condiciones.

**Bucle for**

El bucle `for` nos permite repetir un trozo de código **n veces**. En Python, el bucle for recorre una secuencia (por ejemplo, una secuencia numérica de 1 a n) hasta llegar al último elemento.

En cambio, un número solo no puede usarse en el for, ya que no se puede iterar sobre él.

Así pues, toda variable que pueda ser iterada puede usarse en el bucle for, como las listas, las tuplas, las cadenas de texto o los diccionarios.

```python
# Podemos recorrer listas ya definidas

# Creamos una lista
series_lista = ['Juego de Tronos', 'Mad Men', 'The Witcher']

# Recorremos los elementos de la lista
for s in series_lista:
  # Mostramos el valor s por pantalla en cada iteración
  print(s)
```

```python
# Tuplas
# Creamos una tupla
series_tupla = ('Juego de Tronos', 'Mad Men', 'The Witcher')

# Recorremos los elementos de la tupla
for i in series_tupla:
  # Mostramos el valor i por pantalla en cada iteración
    print(i)
```

```python
# Otros elementos iterables como diccionarios y cadenas de caracteres

# Definimos elementos iterables
series_dict = {"Juego de Tronos": 8, "Mad Men": 7, "The Witcher": 3}
series_str = 'Juego de Tronos'

print("---Diccionario---")
for i in series_dict:
    print(i, end="\n")

print("---Cadena de texto---")
for i in series_str:
    print(i, end=" ")
# Añadimos un espacio (" ") entre las letras
# para que la palabra se muestre en la misma línea
```

También podemos añadir un for dentro de un for, creando lo que se conoce como **for anidado**. Esto puede ser útil si queremos iterar algún objeto que dentro tiene otros elementos que también se pueden iterar.

```python
# Creamos una lista que contiene 2 listas
list_ani = [
    ["Juego de Tronos", "Mad Men", "The Witcher"],
    ["Harry Potter", "Terramar", "Dune"]
            ]

# Mostramos el resultado de recorrer la lista
# con un único for
print("---Único for---")
for n in list_ani:
    print(n)

# Recorremos cada elemento individualmente
# con un for anidado
print("\n---For anidado---")
for n in list_ani:
  for m in n:
    print(m)
```

```python
# Creamos un diccionario con un diccionario dentro
dict_ani = {'series':
                    {"Juego de Tronos": 8, "Mad Men": 7, "The Witcher": 3},
            'libros':
                    {"Harry Potter": 7, "Terramar": 7, "Dune": 6}
            }

# Mostramos el resultado de recorrer el primer diccionario
# con un único for
print("---Único for---")
for n in dict_ani:
    print(n)

# Recorremos cada llave individualmente de los 2 diccionarios
# con un for anidado
print("\n---For anidado---")
for n in dict_ani:
  for m in dict_ani[n]:
    print(m)
```

**Bucle While**

El bucle `while` nos permite ejecutar una parte del código repetidas veces, mientras una condición determinada se cumpla. Cuando ya no se cumpla esta condición, se saldrá del bucle.

```python
from IPython import display
display.Image("images/while-loop.jpeg")
```

```python
# Vamos a usar un bucle while para que se muestre por pantalla
# un mensaje si el valor de x es mayor a 0

# Definimos x con un valor de 4
x = 4

# Ponemos la condición x > 0 en el while
while x > 0:
  # Mostramos el mensaje
  print("El valor de x es mayor que 0: ", x)

  # En cada iteración restamos 1 a x
  x -=1
```

```python
# También se pueden usar condiciones no explícitas

# Usamos la lista de un apartado anterior
lista_while = ["Juego de Tronos", "Mad Men", "The Witcher"]

# No especificamos condición explícita, pero while entiende que
# debe continuar mientras la lista no esté vacía
while lista_while:
    lista_while.pop(0)
    print(lista_while)
```

**Importante**: Se tiene que vigilar mucho con los `while` ya que pueden dar lugar a **bucles infinitos**. Si la condición siempre se cumple, el while no parará nunca, y esta situación raramente es deseable.

**Break y Continue**

 Las instrucciones `break` y `continue` se usan para controlar y alterar el comportamiento de los bucles `for` y `while`.

 - **break** permite interrumpir la ejecución del bucle, salir completamente de él.
 - **continue** permite saltar la iteración actual, sin salir del bucle. Continúa en la siguiente iteración, si la hay.

Ahora vamos a usar el mismo código pero con break or continue, para ver su diferente comportamiento.

```python
## Break:

# Vamos a usar el ejemplo anterior
# En este caso, si x = 1, vamos a saltar interrumpir la ejecución

# Definimos x con un valor de 4
x = 4

# Ponemos la condición x > 0 en el while
while x > 0:
  # En cada iteración restamos 1 a x
  x -=1
  # Cuando se de esta condición
  # se para la ejecución
  if x ==1:
    break
  # Mostramos el mensaje
  print("El valor de x es mayor que 0: ", x)
```

```python
## Break:

# Tambien podemos usar una lista

# Creamos una lista
lista_libros = ["Harry Potter", "Terramar", "Dune"]

# Recorremos los elementos de la lista
for libro in lista_libros:
  # Cuando se encuentre este elemento
  # se para la ejecución
  if libro == 'Terramar':
    break
  # Mostrar elemento por pantalla
  print(libro)
```

```python
## Continue:

# En este caso, si x = 1, vamos a saltar la iteración

# Definimos x con un valor de 4
x = 4

# Ponemos la condición x > 0 en el while
while x > 0:
  # En cada iteración restamos 1 a x
  x -=1
  # Cuando se de esta condición
  # se salta la iteración
  if x ==1:
    continue
  # Mostramos el mensaje
  print("El valor de x es mayor que 0: ", x)
```

```python
## Continue:

# Tambien podemos saltar un elemento de una lista

# Creamos una lista
lista_libros = ["Harry Potter", "Terramar", "Dune"]

# Recorremos los elementos de la lista
for libro in lista_libros:
  # Cuando se encuentre este elemento
  # se salta la iteración
  if libro == 'Terramar':
    continue
  # Mostrar elemento por pantalla
  print(libro)
```

### 2. Funcionalidades avanzadas para iterar: range, enumerate, zip <a name="id2"></a>

En los apartados anteriores hemos visto como controlar el flujo del código con las estructuras condicionales, los bucles y break/continue, así como crear funciones básicas. En esta sección vamos a profundizar en estos temas con herramientas que proporcionan funcionalidades más avanzadas.

#### 2.1. Range

La función `range` devuelve una secuencia de números. Por defecto, empieza desde 0 e incrementa en 1 hasta detenerse antes de un número especificado.

```python
# Usamos range con un valor de 5
# Por defecto usa el valor de inicio 0
# Y el valor incrementar de 1
for n in range(5):
  print(n)
```

Si se prefiere tener más control del inicio y el valor incremental, se pueden especificar esta información cuando llamamos la función range():

`range(inicio, fin, incremento)`

```python
# Vamos a empezar en el 0
# E ir hasta 10 (no incluido)
# Con incrementos de 2
for n in range(2, 10, 2):
  print(n)
```

#### 2.2. Enumerate

La función `enumerate` permite guardar el número de iteraciones en un bucle. Este objeto de enumeración contiene un recuento (desde el principio, que siempre tiene un valor predeterminado de 0) y un valor obtenido al iterar sobre el objeto iterable.

Veamos un ejemplo de su utilidad:

```python
# Vamos a usar una lista de la primera sección
series_lista = ['Juego de Tronos', 'Mad Men', 'The Witcher']

# Queremos mostrar cada elemento de la lista con su iterador
i=0
print("Resultado método manual")
for serie in series_lista:
  print(i, serie)
  i +=1

# Esta es una manera muy manual, no muy "Pythonica"

#Vamos a usar ahora ennumerate
print("Resultado con enumerate")
for i, serie in enumerate(series_lista):
  print(i, serie)

# Obtenemos el mismo resultado pero con la función enumerate
# tiene menos líneas y es más eficiente
```

#### 2.3. Zip

 La función `zip`, combinada con un for, permite iterar múltiples iterables (listas, diccionarios) en paralelo.

```python
# Creamos 3 listas
series_lista = ['Juego de Tronos', 'Mad Men', 'The Witcher']
temporadas_lista = [8, 7, 3]
plataforma_lista = ['HBO', 'Amazon Prime', 'Netflix']

# Usamos zip para obtener los elementos de cada lista en el for
for serie, temp, platf in zip(series_lista, temporadas_lista, plataforma_lista):
    # Mostramos el resultado por pantalla
    print("Puedes encontrar las {} temporadas de la serie {} en {}".format(
        (temp), (serie), (platf)))
```

 Si las listas tienen longitudes diferentes, la iteración se detiene cuando la lista más pequeña se acaba.

```python
# Creamos 3 listas
series_lista = ['Juego de Tronos', 'Mad Men'] # ahora solo tiene 2 elementos
temporadas_lista = [8, 7, 3]
plataforma_lista = ['HBO', 'Amazon Prime', 'Netflix']

# Usamos zip para obtener los elementos de cada lista en el for
for serie, temp, platf in zip(series_lista, temporadas_lista, plataforma_lista):
    # Mostramos el resultado por pantalla
    print("Puedes encontrar las {} temporadas de la serie {} en {}".format(
        (temp), (serie), (platf)))
```

La función `zip` está definida para cualquier objeto iterable. Por lo tanto podemos usarla también con diccionarios.

```python
esp = {'1': 'Uno', '2': 'Dos', '3': 'Tres'}
eng = {'1': 'One', '2': 'Two', '3': 'Three'}

# Para poder ver la clave y el valor del diccionario
# usamos la función items
# y guardamos los dos valores en una tupla
for (k1, v1) in esp.items():
    print(k1, v1)

print("---------------")

# Ahora con zip podemos iterar de manera paralela los 2 diccionarios
# En este caso k1 y k2 tienen el mismo valor, así que solo mostramos uno
for (k1, v1), (k2, v2) in zip(esp.items(), eng.items()):
    print(k1, v1, v2)
```

### 3. List comprehensions  <a name="id3"></a>

Las `list comprehension` permiten crear listas de elementos en una sola línea de código de forma sencilla y elegante sin usar estructuras de control muy complejas.

La sintaxis de las list comprehension es la siguiente:

`lista = [<expresion> for <elemento> in <iterable> <condicion>]`

Vamos a ver unos ejemplos:

```python
# Primero vamos a usar un for,
# para que se quede con los elementos de la lista
# que empiecen con A

# Creamos una lista con nombres
nombres = ['Alex', 'Jose', 'Alvaro', 'Pedro']

# Creamos una 2a lista para guardar el subset
nombres_a = []

# Recorremos los elementos de la lista
for n in nombres:
  # Si el nombre empieza por A
   if n.startswith("A"):
      # Se guarda en la 2a lista
      nombres_a.append(n)

# Mostramos el resultado de la nueva lista
print(nombres_a)
```

Este código es perfectamente válido y correcto, aunque puede parecer un poco largo. Vamos a hacer lo mismo con una `list comprehension`.

```python
# Creamos una lista con nombres
nombres = ['Alex', 'Jose', 'Alvaro', 'Pedro']

# Usamos la list comprehension siguiendo la sintaxi comentada
nombres_a = [n for n in nombres if n.startswith("A")]

# Mostramos el resultado de la nueva lista
print(nombres_a)
```

¡Hemos obtenido el mismo resultado pero en una sola línea!

En el caso de tenir múltiples condiciones (if-else) la estructura cambia un poco de orden:


`lista = [<expresion> if <condicion> else <otra_expresión> for <elemento> in <iterable>]`





```python
# Definimos una lista de números
numeros = [1, 2, 3, 4, 5, 6]

# Usamos list comprehension con if-else
# Si el número es divisible por 2 se mantiene el valor
# En caso contrario, se multiplica por 2
nueva_lista = [x if x % 2 == 0 else x * 2 for x in numeros]

# Mostramos la nueva lista
print(nueva_lista)
```

### 4. Definición de funciones <a name="id4"></a>

#### 4.1. ¿Qué es una función?

Las **funciones** son el método principal para organizar y reutilizar código en Python. Como regla general, si se tiene que repetir el mismo código o uno muy similar más de una vez, es óptimo escribir una función reutilizable. Las funciones también pueden ayudar a que el código sea más legible, ya que encapsulan parte del código y permite reducir el número de líneas en el archivo principal.

Las funciones se declaran con la palabra clave `def` y se devuelven con la palabra clave `return`. Si la función necesita unos parámetros de entrada, como la función suma, se especifican entre paréntesis después del nombre de la función.

Así definimos la función, pero no se ejecuta aún el código de dentro. Para hacerlo se tiene que escribir el nombre de la función y especificar los parámetros de entrada.

```python
# Creamos la función suma
def suma(a, b):
  c = a + b
  return c

# Utilizamos la función para sumar 1 y 2
suma(1,2)
```

También se pueden especificar unos valor por defecto en la función, por ejemplo:

```python
# Creamos la función suma
# con unos valores por defecto
def suma(a=2, b=3):
  c = a + b
  return c

# Ahora podemos no especificar parámetros de inicio
print("Suma params por defecto: ", suma())

# También podemos cambiarlos
# Ahora podemos no especificar parámetros de inicio
print("Suma con params: ", suma(1,2))
```

En el `return` de la función podemos tener múltiples resultados. Por otro lado, si no se especifica `return`, la función devolverá `None`.

```python
# Creamos la función suma y resta
# con unos valores por defecto
def suma_resta(a=2, b=3):
  c = a + b
  d = a - b
  return c, d

# Usamos la nueva función para obtener 2 resultados
c, d = suma_resta()

# Mostramos los resultados por pantalla
print("Resultado de la suma: ", c)
print("Resultado de la resta: ", d)
```

```python
# Definimos una función sin return
def HolaMundo():
  print("Hola Mundo!")

m = HolaMundo()
print("Resultado de la función sin return: ", m)
```

### 5. Gestión de errores y excepciones <a name="id5"></a>

#### 5.1 Errores y excepciones

Considerad la siguiente situación:

Estamos escribiendo un trozo de código, le damos a ejecutar, y Python lanza un error. Estos errores son principalmente de dos tipos:
- Errores de sintaxis: hemos escrito mal alguna palabra o nos hemos dejado algún elemento del código (e.g. un paréntesis, una coma, etc). En estos casos, el intérprete de Python lanza un error e indica en qué línea del código está el problema.

```python
# Escribimos un for sin los dos puntos finales
for n in range(5)
  print(n)

# En el mensaje de error nos indica dónde y qué ha producido el error
```

- Errores de ejecución: son errores detectados durante la ejecución, a pesar de que la sintaxis sea correcta. También se denominan **excepciones**.

```python
# Python no permite hacer una división entre 0
a = 5
b = 0
print(a/b)

# Podéis ver que la excepción es bastante explicativa:
# ZeroDivisionError: division by zero
```

Vamos a ver un ejemplo de cada una:

```python
# Si intentamos usar una variable no declarada
# Tendremos un NameError
print(var_no_declarada)
```

#### 5.2 Gestión de excepciones



Así pues, cada vez que ocurre un error y Python no está seguro de qué hacer a continuación, se crea un objeto de **excepción**. Si el código trata la excepción, el programa continuará ejecutándose. Si no especifica qué hacer cuando ocurre una excepción, el programa se detendrá y mostrará un aviso, que incluye un informe de la excepción que se generó.

Las excepciones pueden ser capturadas y gestionadas adecuadamente usando de las instrucciones `try` y `except`.

Veamos como funcionan con los ejemplos anteriores:

```python
# Ejemplo ZeroDivisionError
# Python no permite hacer una división entre 0
a = 5
b = 0
try:
  print(a/b)
except ZeroDivisionError:
    print("No se puede dividir un número entre 0")

# Con el try-except el programa no se interrumpe
# sino que se captura el error y se muestra el mensaje especificado
```

Si no se sabe que tipo de excepción puede ocurrir, se puede usar la forma genérica `Exception`.

```python
# Definimos dos variables
a = 2
b = 'Hola'

try:
  a/b
except Exception:
  print("Un problema ha ocurrido!")
```

Si queremos un poco más de flexibilidad, podemos combinar el `try-except` con el `else` y el `finally`.

```python
# Usamos el ejemplo anterior
a = 2
b = 'Hola'

try:
    # Forzamos excepción
    c = a/b
except:
    # Se entra ya que ha habido una excepción
    print("Entra en except, ha ocurrido una excepción")
else:
    # No entra en el else, ya que ha ocurrido una excepción
    print("Entra en else, no ha ocurrido ninguna excepción")
finally:
    # También entra porque finally es ejecutado siempre
    print("Entra en finally, se ejecuta el bloque finally")

# Entra en except, ha ocurrido una excepción
# Entra en finally, se ejecuta el bloque finally
```

Con Python se pueden lanzar excepciones específicas con la instrucción `raise`, que permite definir qué tipo de error se genera y el mensaje que se muestra por pantalla.

```python
# Definimos una variable
x = -5

# No queremos que la variable sea negativa
if x < 0:
    raise Exception("[ERROR], el valor tienen que ser positivo")
```
