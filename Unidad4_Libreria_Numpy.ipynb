### 1. Librería Numpy<a name="id1"></a>

#### 1.1. ¿Qué es?

NumPy (Numerical Python) es una librería fundamental en el lenguaje de programación Python para realizar operaciones numéricas y matemáticas de manera eficiente y con un enfoque en matrices y arrays multidimensionales.

Algunas características clave de NumPy incluyen:

- **Arrays multidimensionales:** NumPy proporciona la clase `ndarray` que permite crear arrays n-dimensionales, lo que resulta útil para representar datos tabulares, imágenes, señales, y otros tipos de información multidimensional.

- **Operaciones vectorizadas:** En lugar de iterar sobre cada elemento en un bucle, NumPy permite realizar operaciones sobre arrays directamente (siguiendo la lógica de la algebra de matrices), lo que mejora drásticamente la eficiencia computacional.

- **Indexación avanzada:** Permite indexar y realizar operaciones en subconjuntos de arrays, lo que es esencial para manipular datos de manera eficiente.

- **Broadcasting:** NumPy implementa broadcasting, que permite operaciones entre arrays de diferentes tamaños y formas, evitando la necesidad de expandirlos manualmente.

- **Funciones matemáticas:** NumPy proporciona una amplia variedad de funciones matemáticas y trigonométricas, operaciones lineales álgebra, estadísticas, entre otras.

- **Integración con otras librerías:** NumPy se utiliza en conjunto con muchas otras librerías populares en el ámbito científico y de análisis de datos, como SciPy, Pandas y Matplotlib.

- **Eficiencia:** NumPy está implementado en C y Fortran, lo que le da una gran eficiencia en términos de velocidad de cálculo en comparación con las estructuras de datos nativas de Python.

```python
# Para usar la librería numpy primero tenemos que importarla
# Para facilitar su uso, siempre se abrevia su nombre a np
import numpy as np
```

#### 2.1. Creación de arrays y matrices

En NumPy, el objeto principal consiste en un conjunto multidimensional de números (generalmente) del mismo tipo, llamado `array`.

```python
# Creamos un array 1D
array1d = np.array([1, 2, 3, 4, 5])
print("Array 1D",array1d)

# Creamos un array 2D
array2d = np.array([[1, 2, 3],
                    [4, 5, 6]])
print("Array 2D", array2d)
```

Para obtener las propiedades de las matrices (dimensión, tamaño, número de elementos), podemos usar las funciones siguientes:

**array.ndim**

Indica la dimensión (número de ejes) del array. Por ejemplo, un array 1D tiene dimensión de 1, mientras que un array 2D tiene valor de 2.

**array.shape**

Muestra los elementos de cada dimensión de la matriz. En un array 2D, el primer elemento se refiere a las filas, y el segundo, a las columnas. Por lo tanto, un *shape* de (2,3) indica que el array tiene 2 filas y 3 columnas.

**array.size**

Representa la cantidad total de elementos presentes en la matriz. Siguiendo el ejemplo anterior, un array de dimensiones (2,3), tendrá 2*3 = 6 elementos.

Vamos a poner en práctica estas propiedades con los dos arrays creados anteriormente.

```python
# Mostramos el resultado de estas propiedades
print("ndim de array 1D",array1d.ndim)
print("shape de array 1D",array1d.shape)
print("size de array 1D",array1d.size)
print("-------------------------------")
print("ndim de array 2D",array2d.ndim)
print("shape de array 2D",array2d.shape)
print("size de array 2D",array2d.size)
```

Además de `np.array`, existen otras funciones para crear nuevas matrices. Por ejemplo, las funciones `zero` y `ones` crean matrices de 0 o 1, respectivamente, con una longitud o forma determinada. La función `empty` crea una matriz sin inicializar sus valores a ningún valor en particular. También se puede crear una matriz de números randoms con `random`.

Numpy también permite crear matrices típicas de la algebra lineal, como la **matriz identidad**, con la función `eye`.

```python
print("Array 1D: ")
array_1d_zeros = np.zeros(10)
print(array_1d_zeros.ndim)
print(array_1d_zeros)

print("Array 2D: ")
array_2d_zeros = np.zeros((3, 6))
print(array_2d_zeros.ndim)
print(array_2d_zeros)

print("Array 3D: ")
array_empty_3d = np.empty((2, 3, 2))
print(array_empty_3d.ndim)
print(array_empty_3d)

print("Array Random: ")
arr_rand = np.random.randn(6, 3)
print(arr_rand.ndim)
print(arr_rand)

# La matriz identidad siempre es cuadrada
# indicamos el número de filas/columnas
print("Array Matriz identidad: ")
array_eye = np.eye(3)
print(array_eye.ndim)
print(array_eye)
```

**Importante**: La función `empty` a veces no da exactamente zero, sino números muy próximos (e.g. 4.99052583e-310)

Cuando creamos un array podemos especificar el tipo de los elementos que contiene. Si no lo especificamos, NumPy asumirá el tipo que le parezca más probable.

```python
# Creamos dos arrays de tipo distinto
arr1 = np.array([1, 2, 3], dtype=np.float64)
arr2 = np.array([1, 2, 3], dtype=np.int32)

# Mostramos el resultado por pantalla
print("Array 1",arr1)
print("Tipo de los elementos de array 1", arr1.dtype)
print("Array 2",arr2)
print("Tipo de los elementos de array 2", arr2.dtype)

# También podemos cambiar el tipo
float_arr2 = arr2.astype(np.float64)
print("Array 2 modificado",float_arr2)
print("Tipo de los elementos de array 2 modificado", float_arr2.dtype)
```

Una vez creado un array, podemos modificar el valor de sus elementos accediendo a ellos por su índice.

```python
# Creamos un array de zeros con 8 elementos
array_1d_zeros = np.zeros(8)
print("Array inicial: ",array_1d_zeros)

# Modificamos algunos elementos
array_1d_zeros[0] = 7
array_1d_zeros[5] = 4
array_1d_zeros[-1] = 3

# Mostramos el array modificado
print("Array modificado: ",array_1d_zeros)
```

```python
# Podemos hacer lo mismo con una matriz N-dimensional
# Creamos un array 2D
array_2d_zeros = np.zeros((3, 2))
print("Array inicial:\n ",array_2d_zeros)

# Modificamos algunos elementos
array_2d_zeros[0, 0] = 7
array_2d_zeros[2, 1] = 4
array_2d_zeros[1, 0] = 3

# Mostramos el array modificado
print("Array modificado:\n ",array_2d_zeros)
```

#### 2.2. Slicing e iteración

En NumPy hay diferentes maneras de seleccionar un subconjunto de los datos. Las matrices unidimensionales son simples, ya que funcionan de manera similar a las listas de Python.

```python
from IPython import display
display.Image("images/matrices_slicing.png", width=500)
```

```python
# Creamos un array de 15 elementos
arr1 = np.arange(15)

# Podemos seleccionar una parte del array
# De manera similar a lo que haríamos con una lista
arr2 = arr1[5:11]

# También podemos cambiar el valor de varios elementos a la vez
arr3 = arr1.copy()
# Del sexto elemento al final, cambiamos el valor por 1
arr3[5::] = 1

# Mostramos los resultados
print("Array1:\n",arr1)
print("Subset de array1:\n ",arr2)
print("Subset de array1 modificado:\n ",arr3)
```

```python
# También podemos hacer lo mismo con matrices n-dimensionales
arr2d = np.array([[1, 2, 3, 4], [0, 1, 1, 0]])

# Mostramos el array
print("Array2d:\n",arr2d)

# Hacemos una copia del array 2D para modificarlo
arr2d_mod = arr2d.copy()

# Modificamos el valor del primer elemento
arr2d_mod[0,0] = 100

# Mostramos el array modificado
print("Array2d modificado:\n",arr2d_mod)

# Modificamos el valor de la 1a fila
arr2d_mod[0,:] = -7

# Mostramos el array modificado
print("Array2d modificado:\n",arr2d_mod)

# Nos quedamos con la 1a columna
arr2d_sub1 = arr2d[:, 1]

# Nos quedamos con la 1a fila, de la 2a columna al final
arr2d_sub2 = arr2d[0, 1:]

print("Subset de array2d:\n ",arr2d_sub1)
print("Subset de array2d:\n ",arr2d_sub2)
```

```python
# También podemos iterar y mostrar
# los elementos del array 2D
for i in range(arr2d.shape[0]):
    for j in range(arr2d.shape[1]):
        print("Elemento ", arr2d[i, j])
```

#### 2.4. Operaciones aritméticas con matrices

Las matrices permiten expresar operaciones aritméticas que modifican los elementos de la matriz sin escribir ningún bucle for. Esto se conoce como **vectorización**. Cualquier operación aritmética entre matrices del mismo tamaño aplica la operación por elementos.

Veamos algunos ejemplos:

```python
# Creamos dos arrays 2D
arr1 = np.array([[1, 2, 3], [4, 5, 6]])
arr2 = np.array([[1, 1, 1], [2, 2, 2]])

# Creamos otro array multiplicando arr1 y arr2
arr3 = arr1*arr2

# Creamos otro array sumando los arrays
arr4 = arr1+arr2

# Mostramos todos los arrays
print("Array1:\n",arr1)
print("Array2:\n ",arr2)
print("Array3:\n ",arr3)
print("Array4:\n ",arr4)

# Tambien podemos usar operaciones aritméticas on escalares
# que se aplicarán a todos los elementos de la matrix
arr5 = 3*arr1
arr6 = arr1/2

# Mostramos los nuevos arrays
print("Array5:\n",arr5)
print("Array6:\n ",arr6)

# También podemos aplicar comparaciones
arr7 = arr1 > arr2

# Mostramos el resultado
print("Array7:\n",arr7)
```

Respecto a la multiplicación de matrices, hay de dos tipos:
- Multiplicación elemento por elemento
- Multiplicación de matrices

La multiplicación elemento por elemento multiplica cada elemento de la matriz 1 con el elemento de la misma posición de la matriz 2. Por esta razón, ambas matrices deben tener las mismas dimensiones.

En cambio, en la multiplicación matricial, la primera debe tener el mismo número de columnas que filas la segunda. La matriz resultante del producto quedará con el mismo número de filas de la primera y con el mismo número de columnas de la segunda.

```python
# Creamos 2 matrices 2d
arrA= np.array([[1,2], [0,1]])
arrB = np.array([[1,2], [4,3]])

# Hacemos una multiplicación elemento por elemento
arrC = arrA*arrB

# Si hacemos la multiplicación matricial
arrD = arrA.dot(arrB)

# Mostramos los resultados
print("Multiplicación elemento por elemento:\n",arrC)
print("Multiplicación matricial:\n ",arrD)
```

```python
# Usamos las dos matrices arr1 y arr2
# La multiplicación mostrada anteriormente
# era elemento por elemento
arr3 = arr1*arr2

# Si hacemos la multiplicación matricial
# nos dará error
#arr8 = arr1.dot(arr2)
# ValueError: shapes (2,3) and (2,3) not aligned: 3 (dim 1) != 2 (dim 0)

# Para multiplicarlas de manera matricial
# podemos transponer arr2
arr2b = arr2.T
arr8 = arr1.dot(arr2b) # Ahora si!

print("Dimensiones de arr1", arr1.shape)
print("Dimensiones de arr2b", arr2b.shape)

# Mostramos los resultados
print("Multiplicación elemento por elemento:\n",arr3)
print("Multiplicación matricial:\n ",arr8)
```

#### 2.4. Funciones matemáticas en arrays

NumPy incluye una gran cantidad de operaciones matemáticas que se pueden aplicar a los arrays. Podéis consultar la documentación oficial de la [librería](https://numpy.org/doc/stable/reference/routines.math.html) para más información.

Ahora vamos a ver algunos ejemplos:

```python
# Creamos un array de 10 elementos
arrA = np.arange(10)

# Aplicamos la raiz cuadrada a todos los elementos del array
arrB = np.sqrt(arrA)

# Mostramos los resultados
print("Array inicial:\n",arrA)
print("Array después de la raiz:\n ",arrB)

# Creamos dos arrays random
x = np.random.rand(5)
y = np.random.rand(5)

elem_max = np.max(x)
elem_min = np.max(y)

# Mostramos los resultados
print("Array x:\n",x)
print("Array y:\n",y)
print("Valor máximo de x:\n ",elem_max)
print("Valor mínimo de y:\n ",elem_min)
```
