### Ejercicio 1

En este ejercicio vamos a trabajar las operaciones con matrices.

a) Crea las matrices siguientes:

- Una matriz (A) 4x4 que contenga todos los enteros entre 1 y 16.
- La matriz identidad (B) 4x4.
- Una matriz (C) 5x4 con valores enteros aleatorios (*random*) entre 0 y 9.


b) Considerando las matrices del apartado anterior, haz las siguientes operaciones:

- Multiplica las matrices A y B. Explica el resultado obtenido.
- Multiplica las matrices A y C. Explica brevemente qué método de multiplicación has elegido y el resultado obtenido.

**Nota**: En ambos apartados, muestra el resultado por pantalla para comprobar que las matrices se han creado correctamente.

```python
#a)

# Cargamos la librería NumPy
import numpy as np

# Creamos la Matriz A
A = np.arange(1, 17).reshape(4, 4)

# Creamos la Matriz B
B = np.eye(4)

# Creamos la Matriz C
C = np.random.randint(10, size=(5,4))

# Mostramos el resultado del apartado a)
print('---Matriz A---\n', A)
print('---Matriz B---\n', B)
print('---Matriz C---\n', C)
```

```python
#b)

# Multiplicación matricial de A y B
print("El resultado de A.dot(B) =\n{}".format(A.dot(B)))

# Multiplicación elemento por elemento de A y B
print("El resultado de A*B =\n{}".format(A*B))

# Multiplicación matricial de A y C
print("El resultado de C.dot(A) =\n{}".format(C.dot(A)))
```

**Respuesta**

**Apartado A**

La multiplicación entre las matrices A y B es posible en dos modalidades: multiplicación matricial, que requiere que el número de columnas en A coincida con el número de filas en B, y multiplicación elemento por elemento, que es factible cuando ambas matrices poseen dimensiones idénticas.

En el escenario inicial, tanto A como B se presentan como matrices 4x4. Dado que comparten la misma dimensión, tenemos la flexibilidad de aplicar tanto la multiplicación matricial como la multiplicación elemento por elemento. En ambos enfoques, las matrices resultantes mantendrán una configuración de 4x4.


**Apartado B**

En el segundo escenario, se presenta A como una matriz 4x4 y C como una matriz 5x4. La opción de multiplicación elemento por elemento queda descartada debido a que las dimensiones de ambas matrices no concuerdan de manera precisa. Por lo tanto, solo nos queda la posibilidad de efectuar una multiplicación matricial.

Para que dos matrices A y C puedan ser multiplicadas en un enfoque matricial, es necesario que el número de columnas en A coincida con el número de filas en C. Al intentar llevar a cabo la multiplicación matricial de A·C, se generará un error, ya que el número de columnas en A (4) no se corresponde con el número de filas en C (5). No obstante, es posible realizar la operación C·A sin inconvenientes, debido a que el número de columnas en C (4) es igual al número de filas en A (4).

### Ejercicio 2

Realiza la normalización de una matriz aleatoria de dimensiones 10x10 empleando sus valores de `media` y `desviación estándar`,  tal como se indica en la siguiente equación:

x_norm = (x - media)/std

A continuación, verifica que los valores de media y desviación estándar en la matriz normalizada se aproximan considerablemente a 0 y 1 respectivamente.

**Nota:** Para resolver este ejercicio os pueden ser de utilidad las funciones [mean](https://numpy.org/doc/stable/reference/generated/numpy.mean.html) o [std](https://numpy.org/doc/stable/reference/generated/numpy.std.html) de NumPy.


```python
# Cargamos la libería NumPy
import numpy as np

# Creamos una matriz random de 10x10
M = np.random.random((10,10))

# Calculamos la media y la desviación
M_media = np.mean(M)
M_std = np.std(M)

# Normalizamos la matriz seguiendo la equación del enunciado
M_norm = (M - M_media) / M_std

# Comparamos los valores de la media y la std antes y después de la normalización
# Mostramos que los valores de la media y la std de la matriz normalizada
# son cercanos a 0 y 1, respectivamente
print('Media de la matriz inicial: ', M_media)
print('Desvación estándar de la matriz inicial: ', M_std)
print('Media de la matriz normalizada: ',np.mean(M_norm))
print('Desvación estándar de la matriz normalizada: ', np.std(M_norm))
```

### Ejercicio 3

En este ejercicio pondremos en practica las operaciones con arrays 1D.

a) Crea un array 1D con elementos del 1 al 20.

b) Crea un subset de este array con los 10 primeros elementos.

c) Intercambia el valor del primer y el último valor de este subset.

d) Divide el array inicial por la suma total del subset.

e) Crea un nuevo array con aquellos elementos mayores a 10.

f) Multiplica por 2 aquellos elementos entre 5 y 15 y crea un nuevo array con esta modificación.

```python
# Cargamos la librería NumPy
import numpy as np

#a) Crear array

# Creamos el array 1D con elementos de 1 a 20
array_ini = np.arange(1, 21)

# Mostramos el resultado por pantalla
print("Array inicial: \n", array_ini)

#b) Subset con 20 primeros elementos

# Seleccionamos los 10 primeros elementos del array inicial
array_subset = array_ini[0:10]

# Mostramos el resultado por pantalla
print("Subset de array inicial: \n", array_subset)

#c) Intercambiar primer y úñtimo elemento

# Hacemos una copia del subset para modificarlo
array_subset_mod = array_subset.copy()

# Modificamos el valor del primer y el último elemento
array_subset_mod[0] = array_subset[-1]
array_subset_mod[-1] = array_subset[0]

# Mostramos el resultado por pantalla
print("Subset de array inicial modificado: \n", array_subset_mod)

#d) Divid array por suma subset

# Hacemos una copia del array inicial para modificarlo
array_ini_mod = array_ini.copy()

# Dividimos el array inicial por la suma del subset
array_ini_mod = array_ini_mod / np.sum(array_subset)

# Mostramos el resultado por pantalla
print("Suma del subset: \n", np.sum(array_subset))
print("Array inicial modificado: \n", array_ini_mod)

# e) Crear nuevo array con solo elementos > 10

array_mayor_diez = array_ini[array_ini>10]
print("Array con elementos > 10: \n", array_mayor_diez)

# f) Crear nuevo array con elementos entre 5 y 15 (incluídos) *2
array_mayor_diez = array_ini[(array_ini>=5) & (array_ini<=15)]*2
print("Array modificado entre 5 y 15 *2: \n", array_mayor_diez)
```

### Ejercicio 4

Crea una matriz 15x15 con el borde exterior de valor 1, y todos los elementos interiores con valor 0.

```python
# Cargamos la librería NumPy
import numpy as np

# Creamos una matriz de unos 15x15
Z = np.ones((15,15))

# Todos los elementos que no están en el borde, los ponemos a 0
Z[1:-1,1:-1] = 0

# Mostramos el resultado por pantalla
print(Z)
```

```python
# Tambien se podría hacer de esta manera
# Un poco más lenta

# Creamos una matriz de zeros 15x15
Z = np.zeros((15,15))

# Ponemos los bordes a 1
Z[0,:] = 1
Z[-1,:] = 1
Z[:,0] = 1
Z[:,-1] = 1

# Mostramos el resultado por pantalla
print(Z)
```

```python
# Cargamos la libería de visualización
import matplotlib.pyplot as plt

# Visualizamos el resultado con la función imshow()
plt.imshow(Z)
```

En este ejercicio, vamos a aplicar nuestros conocimientos de estructuras de control, como bucles `for` y condicionales `if`, para completar una matriz en base a ciertas condiciones.

a) Crea una matriz de 10x10 con valor de 1.

b) Rellena la matriz siguiendo las siguientes indicaciones.
Considerando los valores de cada posición  `(i, j)`:
- En la diagonal, el valor será `2i + 3j`
- Para el resto de posiciones, el valor será `(i*j + 2)`.

```python
# Cargamos la libería NumPy
import numpy as np

# Creamos una matriz 4x4 de 1s
m = np.ones((10, 10), dtype=int)

# Recorremos la matriz por filas y columnas
for i in range(m.shape[0]):
    for j in range(m.shape[1]):
    # Si i y j son iguales (diagonal):
        if i==j:
            m[i, j] = 2*i + 3*j
        else:
            m[i, j] = i*j + 2

# Mostramos el resultado por pantalla
print(m)
```
